find out minimum, maximum, q1,q3,iqr to find 5 number summary 


```python
import numpy as np 
ls_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]

min, q1, median, q3,max =np.quantile(ls_marks, [0, 0.25, 0.50, 0.75,1.0])
```

```python
min, q1, median, q3,max
```

```python
iqr=q3-q1
iqr
```

```python
lower_fence= q1-1.5*(iqr)
higher_fence= q3+1.5*(iqr)
print(lower_fence,
higher_fence)
```
